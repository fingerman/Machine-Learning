## 11.4 Machine Learning

#### a.) Link Prediction

The dataset used in this task have been obtained from NetworkRepository.com -> Animal Social Networks (http://networkrepository.com/asn.php). The data consists of six networks, where each one represents social data from wild birds, collected for consecutive 6-day time window. Here I use two subsequent days.   
The aim was, given the Graph for a particular day (G_t0), to predict Links in the following day (G_t1). There similarity measures between pairs of nodes were used. Then those were compared to choose the best one. The resulting AUC scores and the ROC AUC curves based on those similarity measures are shown below.    
The Jaccard and the Adamic-Adar similarity measures outperform the Preferential Attachment criteria on this dataset. Preferential attachment means that the more connected a node is, the more likely it is to receive new links. Maybe that is why this similarity measure does not work well on wild birds, which are organized into different species. Algorithms based on the common neighbor-identification perform better.

```python
import random
from collections import Counter
import pandas as pd
from sklearn import preprocessing
import powerlaw
```

```python
# a.)
fig1 = plt.figure(figsize=(12,8))
for p in [0,0.2,0.4,0.5,0.8,1]:
    G1 = nx.Graph()
    G1.add_edges_from([(0, 1), (1, 2), (2, 0)])

    while len(G1.nodes.values())<= 20000:
        o = random.randrange(len(G1.nodes.values()))
        s = random.randint(0, 1)
        ot = random.choice(list(G1.neighbors(o)))
        if s == 1:
            G1.add_edge(*[len(G1.nodes.values()), o])
        else:
            e =[len(G1.nodes.values()), ot]
            G1.add_edge(*e)
    degrees = sorted([d for n, d in G1.degree()], reverse=True) 
    degree_N = Counter(degrees)
    dgr, n = zip(*degree_N.items())
    hst = pd.DataFrame([dgr,n])
    hst = preprocessing.normalize(hst)
    plt.loglog(hst[0],hst[1], label="p="+str(p))
    plt.title("Normalized Degree Distributions")
    plt.ylabel("N Log")
    plt.xlabel("Log Degree")
    plt.legend()
# b.) + c.)   
    results = powerlaw.Fit(degrees, discrete=True)
    print(results.power_law.xmin)
    print(results.power_law.alpha) # distribution parameters
   # R, p = results.distribution_compare('power_law', 'lognormal_positive')
   # print(R, p) # try test also to against distributions from the library


```

```python
from sklearn.metrics import roc_curve, roc_auc_score, accuracy_score
import random
```

```python
#a.) Link Prediction

G_t0=nx.read_edgelist("wildbird-network-2.edges", nodetype=int, data=(('weight', float),))
G_t1=nx.read_edgelist("wildbird-network-3.edges", nodetype=int, data=(('weight', float),))
```

```python
print(nx.info(G_t0))
nx.draw(G_t0)
```

```python
print(nx.info(G_t1))
nx.draw(G_t1)
```

```python
n_Gt1 = G_t1.number_of_nodes()
m_Gt1 = G_t1.number_of_edges()
edge_subset = G_t1.edges - G_t0.edges
```

```python
edge_subset_size = G_t0.number_of_edges()
print("Number of edges removed   :", edge_subset_size)
print("Number of edges remaining :", (m_Gt1 - edge_subset_size))
```

```python
#Make prediction using Jaccard Coefficient
pred_jaccard = list(nx.jaccard_coefficient(G_t0))
score_jaccard, label_jaccard = zip(*[(s, (u,v) in edge_subset) for (u,v,s) in pred_jaccard])
# Compute the ROC AUC Score
fpr_jaccard, tpr_jaccard, _ = roc_curve(label_jaccard, score_jaccard)
auc_jaccard = roc_auc_score(label_jaccard, score_jaccard)

# Prediction using Adamic Adar 
pred_adamic = list(nx.adamic_adar_index(G_t0))
score_adamic, label_adamic = zip(*[(s, (u,v) in edge_subset) for (u,v,s) in pred_adamic])
# Compute the ROC AUC Score
fpr_adamic, tpr_adamic, _ = roc_curve(label_adamic, score_adamic)
auc_adamic = roc_auc_score(label_adamic, score_adamic)

# Compute the Preferential Attachment
pred_pref = list(nx.preferential_attachment(G_t0))
score_pref, label_pref = zip(*[(s, (u,v) in edge_subset) for (u,v,s) in pred_pref])
# Compute the ROC AUC Score
fpr_pref, tpr_pref, _ = roc_curve(label_pref, score_pref)
auc_pref = roc_auc_score(label_pref, score_pref)
```

```python
fig3 = plt.figure(figsize=(12,8))
ax = plt.axes()
ax.plot(fpr_jaccard, tpr_jaccard, label='line 1')
ax.plot(fpr_adamic, tpr_adamic, label='line 2')
ax.plot(fpr_pref, tpr_pref, label='line 3')
ax.set_ylabel('True positive rate')
ax.set_xlabel('False positive rate')
ax.set_title("ROC AUC Curve")
ax.legend(('Jaccard Coefficient - AUC: ' + str(auc_jaccard)[:5], 
           'Adamic Adar - AUC: ' + str(auc_adamic)[:5],
           'Preferential Attachment - AUC: ' + str(auc_pref)[:5]),loc =  "lower right")

plt.show()
```

```python
random.seed(5)
proportion_nodes = 0.7
labeled_nodes_iip = random.sample(G_iip.nodes(), int(proportion_nodes * G_iip.number_of_nodes()))

known_labels_iip = np.zeros(n_iip)

nodes_iip = list(G_iip.nodes())


for e in labeled_nodes_iip:
    known_labels_iip[nodes_iip.index(e)] = int(labels_iip[nodes_iip.index(e)])

known_labels_iip = known_labels_iip.astype(int)  
```

```python
plt.figure(figsize=(12,8))
nx.draw(G_iip, node_color = known_labels_iip)
plt.gca().collections[0].set_edgecolor("#000000") # set node border color to black
plt.show()
```

```python
alpha = 0.7
L_sym = nx.normalized_laplacian_matrix(G_iip)

Y = np.zeros((n_iip,2))
Y[known_labels_iip==2, 0] = 1
Y[known_labels_iip==1, 1] = 1
I = np.identity(n_iip)

# Create the F-pred matrix
F_pred = np.linalg.inv(I*(1-alpha) + L_sym) * Y
# Identify the prediction as the argmax
pred = np.array(np.argmax(F_pred, axis=1)*2-1).flatten()
# Compute the accuracy score
succ_rate = accuracy_score(labels_iip, pred)
```
