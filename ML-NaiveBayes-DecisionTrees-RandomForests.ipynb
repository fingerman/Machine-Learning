### 1.1 Recode Variables

In our original dataset the feauture names are strings. To work with the data we need to recode them using proper
and comparable scale level. The recoding starts with "0" at the lowest level of our ordinal variables moving ahead. 

    parents        0 = usual, 1 = pretentious, 2 = great_pret
    has_nurs       4 = proper, 3 = less_proper, 2 = improper, 1 = critical, 0 = very_crit
    form           3 = complete, 2 = completed, 1 = incomplete, 0 = foster
    children       1, 2, 3, 4 = more
    housing        1 = less_conv, 0 = critical
    finance        1 = convenient, 0 = inconv
    social         2 =  non-prob, 1 = slightly_prob, 0 = problematic
    health         2 = recommended, 1 = priority, 0 = not_recom
    


```python
def parents_to_int(series):
    if series == "usual":
        return 0
    elif series == "pretentious":
        return 1
    elif series == "great_pret":
        return 2

nursery_data['parents'] = nursery_data['parents'].apply(parents_to_int)
```

```python
def nurs_to_int(series):
    if series == "proper":
        return 4
    elif series == "less_proper":
        return 3
    elif series == "improper":
        return 2
    elif series == "critical":
        return 1
    elif series == "very_crit":
        return 0

nursery_data['has_nurs'] = nursery_data['has_nurs'].apply(nurs_to_int)
```

```python
def form_to_int(series):
    if series == "complete":
        return 3
    elif series == "completed":
        return 2
    elif series == "incomplete":
        return 1
    elif series == "foster":
        return 0

nursery_data['form'] = nursery_data['form'].apply(form_to_int)
```

```python
nursery_data['children'].replace('more', 4, inplace=True)
```

```python
def housing_to_int(series):
    if series == "convenient":
        return 2
    elif series == "less_conv":
        return 1
    elif series == "critical":
        return 0
    
nursery_data['housing'] = nursery_data['housing'].apply(housing_to_int)   
```

```python
def finance_to_int(series):
    if series == "convenient":
        return 1
    elif series == "inconv":
        return 0
    
nursery_data['finance'] = nursery_data['finance'].apply(finance_to_int)   
```

```python

def social_to_int(series):
    if series == "nonprob":
        return 2
    elif series == "slightly_prob":
        return 1
    elif series == "problematic":
        return 0
    
nursery_data['social'] = nursery_data['social'].apply(social_to_int)      
```

```python
def health_to_int(series):
    if series == "recommended":
        return 2
    elif series == "priority":
        return 1
    elif series == "not_recom":
        return 0
    
nursery_data['health'] = nursery_data['health'].apply(health_to_int) 
```

## 2. Machine Learning Algorithms

In this section we will apply algorithms to make best possible predictions for the classification of new observations. We mentioned that our target variable has four classes. Hence we will apply algorithms which are capable of hndling multiple classes directly. Those will be naive Bayes Classifiers, Decision Trees and Random Forests. We will not go deeper into the theory of how those classification algorithms work. 

### 2.1 Prepare the Data for ML

Generally most of the Machine Learning algorithms do not perform well, when the inmut numerical attributes 
have very different scales. In our data we do not have this case e.g. range from 1 to 10 000 for instance. Algorithms 
which explore distances, like SVM or k-NN, or/and seek to reduce dimentions - PCA, FA etc.,  are sensitive to feature transformation. However it will be not worked with them 
but rather just with graphic-model based classifiers, which are invariant to feature scaling. 

However we need to split the data into training and testing data sets. We use standard 70% / 30% split. This is a classification problem, so stratify the split is stratified according to the target variable - "evaluation".

```python
nursery_attributes = nursery_data.drop("evaluation", axis = 1) 
nursery_target = nursery_data[["evaluation"]]
```

```python
nursery_attributes_train, nursery_attributes_test, nursery_target_train, nursery_target_test = train_test_split(nursery_attributes, nursery_target, stratify=nursery_target, train_size=0.7, test_size=0.3)
```

### 2.2 ML Algorithms

In this report we will no go deeper in the theoretical way the different algorithms function and what role the hyperparameters do. However our aim is of course to find the best predictive algorithms. The classification accuracy is considered as proper measure for the error because the target variable resembles a uniform discrete distribution. 

#### 2.2.1 Naive Bayes Classifiers

Naive Bayes models are considered as efficient, as they learn parameters by looking at each feature individually and collect simple per-class statistics from each feature. There are basically three kinds of naive classifiers implemented in python and  which we can use: `GaussianNB`, `BernoulliNB` and `MultinomialNB`. We apply the last one, as it assumes discrete count data, represented as integers. This model has a single parameter `àlpha`, which controls for the model complexity. 
    Naive Bayes calculates the posterior probability P(c|x) from P(c), P(x) and P(x|c), 

\begin{equation*}
P(c| x)   = \frac{P(x|c)P(c)}{P(x)}
\end{equation*}

where:

$P(c|x)$ is the posterior probability of class (c, target) given predictor (x, attributes). <br />
$P(c)$ is the prior probability of class. <br />
$P(x|c)$ is the likelihood which is the probability of predictor given class. <br />
$P(x)$ is the prior probability of predictor. <br />

Assumptions:
* assumption of independent predictors. In real life, it is almost impossible that we get a set of predictors which are completely independent.    
* It perform well in case of categorical input variables compared to numerical variable(s). For numerical variable, 
    normal distribution is assumed (bell curve, which is a strong assumption).  
* If categorical variable has a category (in test data set), which was not observed in training data set, then model
    will assign a 0 (zero) probability and will be unable to make a prediction. This is often known as “Zero Frequency”.
    To solve this, we can use the smoothing technique. One of the simplest smoothing techniques is called Laplace estimation.

```python
# naive bayes
bayes_clf = MultinomialNB()
bayes_clf.fit(nursery_attributes_train, nursery_target_train['evaluation'])
y_pred = bayes_clf.predict(nursery_attributes_test)
accuracy_score(nursery_target_test, y_pred)
```

#### 2.2.2 Decision Tree Classifier

With decision trees we estimate the probability that an instance belongs to a particular class. They are trained using the CART algrorithm - a nonparametric tool which splits the training set in two subsets using a singule feature and a treshold. Let us denote that featrure $k$ and the threshold $t_k$. It searches for a pair ($k$, $t_k$) that produces the purest subsets. The cost function that the algorithm tries to minimize is:

\begin{equation*}
J(k, t_k)   = \frac{m_{left}}{m}G_{left}+\frac{m_{right}}{m}G_{right}
\end{equation*}

where:  
    $G_{left/right}$ measures the impurity of the left/right subset<br />
    $m_{left/right}$ is the number of instances in the left/right subset
    

The impurity is measured by default by the Gini measure. We say a node is "pure" if all training instances it applies to 
belong to the same class. It can be stated as:  
    \begin{equation*}
G_i   = 1 -  \sum_{k=1}^{n} p_{i,k}^2
\end{equation*}
where:  

$p_{i,k}$ is the ration of klass $k$ instances among the training instances in the $i$-th node.

An alternative measure is the Entrophy. 

Assumptions:
* do not require feature scaling or centering at all

```python
# dec tree
tree_clf = DecisionTreeClassifier(max_depth=5, random_state=0)
tree_clf.fit(nursery_attributes_train, nursery_target_train['evaluation'])
y_pred = tree_clf.predict(nursery_attributes_test)
accuracy_score(nursery_target_test, y_pred)
```

#### 2.2.3 Random Forest Classifier

We will just mention that a random forest is an ensemble of Decision Trees. The algorithm introduces extra randomness when growing trees. Insted of searching for the very best feature when splitting a node, it searches for the best feature among a random subset of features. This results in greater tree diversity and in this way higher bias is traded for lower variance. 

```python
# random forests
forest_clf = RandomForestClassifier(max_depth=5, random_state=0)
forest_clf.fit(nursery_attributes_train, nursery_target_train['evaluation'])
y_pred = forest_clf.predict(nursery_attributes_test)
accuracy_score(nursery_target_test, y_pred)
```

### 2.3 Tuning the models

#### 2.3.1 Cross-Validation


To achieve better evaluation of the model we will use cross-validation. That is a random split into a number (we will choose 7) of different subsets. Those are called also folds and we will perform stratified K-fold cross validation. After the split the method trains and evaluates the model  times, taking a different fold for evaluation every time and training on the rest 6.

The result is an array containing 7 evaluation scores.The cross-validation allows us to get not only an estimate of the model's performance and also a measure of how precise this estimate is (the standard deviation).


```python
bayes_scores = cross_val_score(bayes_clf, nursery_attributes_train, nursery_target_train['evaluation'], cv=7)
tree_scores = cross_val_score(tree_clf, nursery_attributes_train, nursery_target_train['evaluation'], cv=7)
forest_scores = cross_val_score(forest_clf, nursery_attributes_train, nursery_target_train['evaluation'], cv=7)
```

```python
def display_scores(scores):
    print("Scores:", scores)
    print("Mean:", scores.mean())
    print("Standard deviation:", scores.std())
    print("\n")
```

```python
print("Naive Bayes Classifier scoring:")
display_scores(bayes_scores)

print("Decision Tree Classifier scoring:")
display_scores(tree_scores)

print("Random Forest Classifier scoring:")
display_scores(forest_scores)
```

Cross validations provides an estimate of the performance of our model. Right now the Random Forest Classifier model outperforms the others.

#### 2.3.2 Grid Search

We will optimise the hyperparameters of our models. The Grid-Search method allows us to 
specifiy the values of the hyperparamethers we want to try out. Here we evaluate all combinations using cross validation.

```python
# grid search naive bayes

param_grid = [
    {'alpha': [1, 0.1, 0.01, 0.001, 0.0001, 0.00001]}
]
bayes_classifier = MultinomialNB()
grid_search = GridSearchCV(bayes_classifier, param_grid, cv=7, scoring = "accuracy")
grid_search.fit(nursery_attributes_train, nursery_target_train['evaluation'])
best_bayes_classifier = grid_search.best_estimator_
print("Naive Bayes - best parameters:", grid_search.best_params_)
print("Naive Bayes - best score:", grid_search.best_score_)
```

```python
# grid search decision tree
param_grid = [
    {'max_depth': [1, 5, 7, 15, 20], 
     'min_samples_leaf':[2, 5, 10, 12], # regularization
     'max_leaf_nodes':[5, 10, 20]}
]
decis_tree_classifier = DecisionTreeClassifier(random_state=0)
grid_search = GridSearchCV(decis_tree_classifier, param_grid, cv=7, scoring = "accuracy")
grid_search.fit(nursery_attributes_train, nursery_target_train['evaluation'])
best_tree_classifier = grid_search.best_estimator_
print("Decision tree - best parameters:", grid_search.best_params_) 
print("Decision tree - best score:", grid_search.best_score_)
```

```python
# grid search random forest

param_grid = [
    {'max_depth': [5, 7, 15, 20], 
     'n_estimators':[100, 200, 300]   # N of trees
    }]
rand_forest_classifier = RandomForestClassifier(random_state=0)
grid_search = GridSearchCV(rand_forest_classifier, param_grid, cv=7, scoring = 'accuracy')
grid_search.fit(nursery_attributes_train, nursery_target_train['evaluation'])
best_random_forest_classifier = grid_search.best_estimator_
print("Random forest - best parameters:", grid_search.best_params_)
print("Random forest - best score:", grid_search.best_score_)
```

The hyperparameters tuning payed off. We could select substantially better models for the Decision Tree and Random Forest Classifiers. But let us compare the testing scores:

```python
bayes_classifier_score = best_bayes_classifier.score(nursery_attributes_test, nursery_target_test)
tree_classifier_score = best_tree_classifier.score(nursery_attributes_test, nursery_target_test)
forest_classifier_score = best_random_forest_classifier.score(nursery_attributes_test, nursery_target_test)
```

```python
print("Testing scores:")
print("Naive Bayes:", bayes_classifier_score)
print("Decision tree:", tree_classifier_score)
print("Random forest:", forest_classifier_score)
```

After specifying and testing three different algorithms, cross-validating them and tuning their parameters we have found a suitable 
model usind the Random Forest Classifier. This renders a 0.99 accuracy score of the classification when tested with new values.
